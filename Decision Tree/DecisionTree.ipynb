


#  Decision Tree



We import the required packages and the data structure

```python
#importing packages
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeRegressor,plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
#importing data frame
melbourne_file_path='/content/melb_data.csv'
melbourne_data=pd.read_csv(melbourne_file_path)
melbourne_data.columns
```

Training the model and calculating the error

```python
#calling the model,splitting the data,predicting the data and calculating the error
melbourne=melbourne_data.dropna(axis=0)
melbourne_features=['Rooms', 'Bathroom', 'Landsize', 'BuildingArea',
                        'YearBuilt', 'Lattitude', 'Longtitude']
X=melbourne[melbourne_features]
y=melbourne.Price
melbourne_model=DecisionTreeRegressor()
train_X,val_X,train_y,val_y=train_test_split(X,y,random_state=0)
melbourne_model.fit(train_X,train_y)
prediction=melbourne_model.predict(val_X)
error=mean_absolute_error(val_y,prediction)
print(error)
```

Graphical representation of the tree

```python
#plot the tree
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
plot_tree(melbourne_model,filled=True, feature_names=X.columns, class_names=y)
plt.show()
```
