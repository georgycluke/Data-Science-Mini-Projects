# **Simple** **Linear** **Regression**
In this worksheet we will write a program demonstrating simple linear regression.


First we import the required packages for the program and plotting

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error
```

Importing the data frame and seeing the first five rows


```python
df=pd.read_csv('/content/for_beginners_simple_linear_regression_model.csv')
```

```python
df.head()
```

```python
#checking whether there are any null values in data structure
df.isna().any()
```

```python
#sum up the null values in each column separately
df.isna().sum()
```

```python
df.dropna(inplace = True)
```

```python
#splitting the data set into training and testing the data
X_train,X_test,Y_train,Y_test=train_test_split(df[['x']],df['y'],random_state=0)
#Calling out the model,fit the data, predict on X_test,calculate errors
model=LinearRegression()
model.fit(X_train,Y_train)
Y_pred=model.predict(X_test)
r2=r2_score(Y_test,Y_pred)
mean_error=mean_squared_error(Y_test,Y_pred)
print('r2 square error:',r2, '\n')
print('mean square error:',mean_error )
```

```python
#plotting the regression line
Y_pred_train=model.predict(X_train)
plt.scatter(X_train, Y_train, label='Actual data')
plt.plot(X_train, Y_pred_train, color='red', label='Regression line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression')
plt.show()
```
